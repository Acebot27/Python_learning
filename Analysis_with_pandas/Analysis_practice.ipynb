### Min Wage Analysis

```python
# load the data
df = pd.read_csv('data/Minimum Wage Data.csv', encoding='latin') # best encoding after utf-8 is latin
df.head()
```

```python
df.to_csv("data/minwage.csv", encoding="utf-8")
df = pd.read_csv("data/minwage.csv")
df.head()
```

```python
# group the data by state
gb = df.groupby("State")
gb.get_group("Alabama").set_index("Year").head()
```

```python
# create a dictionary of states
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Department.Of."
                                                "Labor.Cleaned."
                                                "Low.Value.2020.Dollars"]].rename(columns={"Department.Of"
                                                                                           ".Labor.Cleaned.Low."
                                                                                           "Value.2020.Dollars": name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Department.Of."
                                                                  "Labor.Cleaned."
                                                                  "Low.Value."
                                                                  "2020.Dollars"]].rename(columns={"Department.Of."
                                                                                                   "Labor.Cleaned."
                                                                                                   "Low.Value."
                                                                                                   "2020.Dollars":name}))
```

```python
# check the data
act_min_wage.head(7)
```

```python
# check the data
act_min_wage.describe()
```

```python
# Checking for correlation
act_min_wage.corr().head()
```

##### Observations
- Saw that there are some problematic data in the form of 0.0 values which gives NaN values in the correlation matrix
- Also saw that there are some states which have no data at all, like Alabama , Tennessee etc.
- Also saw that there are some states which have data for only a few years, like Arkansas, Delaware etc.
- So, we need to clean the data by removing the NaN values and also removing the states which have no data at all.

```python
# finding records with 0.0 for LOW.VALUE.2020.DOLLARS in our main dataframe
issue_df = df[ df['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars']==0 ]
issue_df.head() # problematic data
```

```python
# get the states from the problematic dataframe
issue_df['State'].unique()
```

```python
# dropping the states from wages dataframe which have no data at all
act_min_wage.replace(0, np.NaN).dropna(axis=1).corr().head() # axis=1 to drop columns
# replacing 0 with NaN and then dropping the columns with NaN values
```

```python
# saving the cleaned correlation matrix dataframe
min_wage_corr = act_min_wage.replace(0, np.NaN).dropna(axis=1).corr()
```

```python
# checking existence identified problem still exists after dropping the columns or not
for problem in issue_df['State'].unique():
    if problem in min_wage_corr.columns:
        print("We're missing something here...", problem)
```

```python
grouped_issues = issue_df.groupby("State")

for state, data in grouped_issues:
    if data['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'].sum() != 0.0:
        print("We missed it", state)
```

### Combining datasets and finding relationship between them
- We will be using the dataset of unemployment rate
- We will be using the dataset of Presidential 16 election results
- We will be using the dataset of minimum wage

```python
# unemployment rate dataset
unemp_county = pd.read_csv("./data/ue_us/output.csv")
unemp_county.head()
```

```python
# minimum wage dataset
min_wage = pd.read_csv("./data/minwage.csv", index_col=0)
min_wage.head()
```

```python
# cleaning the minimum wage dataset
min_wage = min_wage.replace(0, np.NaN).dropna(axis=1)
min_wage.head()
```

```python
# get minimum wage for each year and state
def get_min_wage(year, state):
    try:
        return min_wage.loc[year][state]
    except:
        return np.NaN

get_min_wage(2012, "Colorado")
```

```python
# add a new column to the unemployment rate dataset
unemp_county['min_wage'] = list(map(get_min_wage, unemp_county['Year'], unemp_county['State']))
unemp_county.head()
```

```python
# tail of the unemployment rate dataset
unemp_county.tail()
```

```python
# check the correlation between unemployment rate and minimum wage
unemp_county[ ['Rate', 'min_wage'] ].corr()
```

```python
# check the covariance between unemployment rate and minimum wage
unemp_county[ ['Rate', 'min_wage'] ].cov()
```

```python
# Presidential 16 election results dataset
pres16 = pd.read_csv("./data/pres16results.csv")
pres16.head()
```

```python
# saving states and their abbreviations
dfs = pd.read_html("https://www.infoplease.com/state-abbreviations-and-state-postal-codes")
for df in dfs:
    print(df.head())
```

```python
# read the state abbreviations dataset
state_abbv = pd.read_csv("./data/state_abbv.csv", index_col=0)
state_abbv.head()
```

```python
# merge the unemployment rate dataset with the state abbreviations dataset
abbv_dict = state_abbv.to_dict()['Postal Code']
abbv_dict
```

```python
# reduce the range of the unemployment rate dataset
county_2015 = unemp_county[ (unemp_county['Year']==2015) & (unemp_county['Month']=="February") ]
county_2015.head()
```

```python
# renaming columns in the presidential 16 election results dataset
pres16.rename(columns={"county":"County", "st":"State"}, inplace=True)
pres16.head()
```

```python
# set the index for both the datasets (county_2015 and pres16)
for df in [county_2015, pres16]:
    df.set_index(["County", "State"], inplace=True)
county_2015.head()
```

```python
pres16.set_index(["County", "State"], inplace=True)
```

```python
pres16.head()
```

```python
# acquiring only Donald Trump and Hillary Clinton votes from the presidential 16 election results dataset
pres16 = pres16[(pres16['cand']=="Donald Trump") | (pres16['cand']=="Hillary Clinton")]
pres16 = pres16[['cand', 'pct']]
pres16.head()
```

```python
pres16.dropna(inplace=True)
pres16.head()
```

```python
all_together.shape
```

```python
all_together.dropna(inplace=True)
all_together.head()
```

```python
all_together.drop("Year", axis=1, inplace=True)
all_together.head()
```

```python
donald_votes = all_together[ (all_together['cand']=="Donald Trump") & (all_together['pct'] > 0.0) ]
donald_votes.head()
```

```python
hillary_votes = all_together[ (all_together['cand']=="Hillary Clinton") & (all_together['pct'] > 0.0) ]
hillary_votes.head()
```

```python
# correlation between unemployment rate, minimum wages and Donald Trump votes
donald_votes.corr()
```

```python
# correlation between unemployment rate, minimum wages and Hillary Clinton votes
hillary_votes.corr()
```

```python
# covariance between unemployment rate, minimum wages and Donald Trump votes
donald_votes.cov()
```

```python
# covariance between unemployment rate, minimum wages and Hillary Clinton votes
hillary_votes.cov()
```

```python
plt.figure(figsize=(8,8))
labels = [c for c in hillary_votes.corr().columns] # first two letters of the column names
ax = sns.heatmap(hillary_votes.corr(), xticklabels=labels, yticklabels=labels, annot=True, cmap="RdYlGn")
plt.show()
```

##### Conclusion:
- Curiously, min_wage appears to have a negative correlation with the pct vote for Trump, so as minimum wage rises, people are less likely to vote for Trump, for example.
- That makes sense since higher minimum wages are typically a Democratic agenda, though it is fairly curious that, while correlated, they do not attempt to vary together.
- Mostly, correlation is preferred over covariance because it does not get affected by the change in scale.
- Correlation tells you how strongly two variables are related to each other, whereas covariance tells you how and till what extent they are related to each other.
